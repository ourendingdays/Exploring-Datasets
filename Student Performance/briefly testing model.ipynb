```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.preprocessing import StandardScaler
```

```python
STUDENTS = pd.read_csv("student/student-por.csv", sep=';') 

STUDENTS.head()
```

```python
df = STUDENTS
```

```python
df.shape #Check the shape of the dataframe
```

```python
df["G"] = (df["G1"] + df["G2"] + df["G3"])/3

#Attributes to test
df=df[['studytime', 'failures', 'Dalc', 'Walc', 'traveltime', 'freetime', 'Medu', 'Fedu','G']]   
```

```python
df
```

```python
x=np.array(df.drop(['G'],1))
y=np.array(df['G'])   
```

```python
x
```

```python
y.shape
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.33) 
```

```python
from sklearn.linear_model import SGDRegressor

sgd_reg = SGDRegressor(penalty="l2") # specifying Ridge Regression

sgd_reg.fit(x_train, y_train)

accuracy=sgd_reg.score(x_test,y_test)  
print(accuracy)
```
